# Drop unnecessary columns

```python
columns_to_drop = ['EmployeeCount', 'StandardHours']
data = data.drop(columns=columns_to_drop)

print("Columns after dropping unnecessary ones:")
print(data.columns)
```

# Removing duplicate rows

```python
print(f"Number of duplicate rows: {data.duplicated().sum()}")
data = data.drop_duplicates()

print(f"Number of rows after removing duplicates: {len(data)}")
```

# Sanitizing Specific Columns

## Renaming columns for clarity

```python
data = data.rename(columns={
    'ï»¿Age': 'Age',
    'DailyRate': 'daily_rate',
    'DistanceFromHome': 'distance_from_home',
    'JobSatisfaction': 'job_satisfaction',
    'YearsWithCurrManager': 'years_with_manager'
})

print("Columns after renaming:")
print(data.columns)
```

## Standardizing text Entries in Categorical Columns

```python
# Standardize text in 'BusinessTravel' column
data['BusinessTravel'] = data['BusinessTravel'].str.strip().str.replace('_', ' ')

# Convert binary categorical column 'Attrition' to numerical values
data['Attrition'] = data['Attrition'].map({'Yes': 1, 'No': 0})

print(data[['BusinessTravel', 'Attrition']].head())
```

## Handling Invalid Numeric Data

```python
numeric_columns = ['daily_rate', 'distance_from_home']

for col in numeric_columns:
    print(f"Checking for invalid values in {col}:")
    print(data[col][data[col] < 0])
    data[col] = data[col].clip(lower=0)  # Replacing negative values with 0

print(data[numeric_columns].describe())
```

# Checking and Removing NaN values

```python
print("Nan values per column before cleanup:")
print(data.isnull().sum())

data = data.dropna()

print(f"Number of rows after removing NaN values: {len(data)}")
print("Missing values per column after cleanup:")
print(data.isnull().sum())
```
